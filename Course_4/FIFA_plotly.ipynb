```python
import pandas as pd
import plotly as pt
import plotly.offline as offline
import plotly.graph_objs as go
```

```python
df = pd.read_csv("CompleteDataset.csv")
```

```python
def pos_updater(x):
    try:
        a = x.split(" ")
        return a[0]
    except IndexError:
        return x
```

```python
nation = ["Germany"] 
data = df[(df.Nationality.isin(nation))][["Name", "Age", 'Nationality', 'Value', 'Wage', "Club"]]
data['salary'] = data['Wage'].apply(lambda x: int(x[1:-1]+"000"))
data['upd_Value'] = data['Value'].apply(lambda x: x[1:])
data['upd_Value'] = data['upd_Value'].apply(lambda x: float(x[:-1])*1000 if x[-1]=="K" else float(x[:-1])*1000000)
data.sort_values("salary", inplace=True)
data.tail(10)
```

## Гистограмма ценности игроков укажет как меняется зарплата немецких игроков с повышением их рыночной стоимости

```python
trace = go.Scatter(x=data.salary, y=data.upd_Value)

datas = [trace]
layout = go.Layout(barmode = "group", title='German players info',
                   xaxis= dict(title= "Player's Salary",ticklen= 5,zeroline= False), 
                   yaxis= dict(title= "Player's Value",ticklen= 5,zeroline= False))
fig = go.Figure(data = datas, layout = layout)

offline.iplot(fig, validate=False)
```
